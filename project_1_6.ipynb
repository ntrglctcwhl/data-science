# Ranking Customers with RFM: Simple, Powerful, but Just the Beginning

### Recency
To determine **recency** of customer purchases given their name, we join the `customers` and `sales` tables on matching `customer_id`.  
We will group by `customer_id`, and use the `MAX()` function to find the most recent date under each `customer_id`.  
This indicates the recency of purchase and can be denoted as **`customer_recency`**.

---

### Frequency
To display the **frequency** (within a specified date range) of customer purchases, we join the `customers` and `sales` tables on matching `customer_id`.  
Then, grouping by `customer_id`, we calculate `COUNT(*)` of `sale_date` where the date is within the specified range (greater than a chosen **min date**, and less than a chosen **max date**).  
The timeframe can be determined by the data science team as needed.  
This can be denoted as **`customer_frequency`**.

---

### Monetary Value
To display the **average monetary value spending** of each customer, we join the `customers` and `sales` tables on matching `customer_id`.  
From the `sales` table, we consider `total_amount`, which is the total amount of a single sale to that customer.  
Grouping by `customer_id`, we take the **sum of `total_amount`** and display the results as **`customer_spending`**.  
This will give us the total amount of money that the customer has put into the store.

---

### Scoring and Limitations
For each metric, we will **rank customers** (using `RANK()`) and give them 3 separate scores ranging from 1 to the total number of customers, denoted as:

- **`recency_score`**  
- **`frequency_score`**  
- **`spending_score`**

The smaller the score, the better the customer is in that metric.  

Each RFM metric is considered equally important in determining a good customer (since there is no elaboration provided on which metrics the organization considers more heavily).  
Therefore, the **average of these scores** will be taken as a customerâ€™s cumulative score, denoted as **`customer_score`**.

While useful, the RFM model has limitations since it does not account for factors such as **customer age, financial status, demographics, or long-term profitability**.  
It should therefore be considered a **starting point** for segmentation, which can be enriched with additional **behavioral, demographic, or predictive data** for a more complete view of customer value.
